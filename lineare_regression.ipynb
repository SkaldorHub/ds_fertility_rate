```python
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
import os

# create output folder
os.makedirs("diagram_output", exist_ok=True)

country = "Sweden"

# Read fertility Data
df = pd.read_csv("female_laborforce.csv")
output = pd.read_excel("fertility_rates.xls", sheet_name="Data", index_col=0, skiprows=3)
#country_data = output.loc[country]

df_world = df[df["Entity"] == "World"] #way for ourworldindata data

ratio_column = "Ratio of female to male labor force participation rate (%) (modeled ILO estimate)"
ratios = df_world[ratio_column].tolist()

available_years = sorted(df_world["Year"].unique())
years_every_10 = available_years[::10]

x = []
y = []

countries = ["Sweden", "Niger", "China", "Brazil", "Korea, Rep."]
for year in available_years:
    year_str = str(year)
    if year_str not in output.columns:
        continue
    for country in countries: #output.index
        value = output.at[country, year_str]
        if pd.notna(value):
            x.append(year)
            y.append(value)

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel("Year")
ax1.set_ylabel("Female Labor Participation Ratio", color=color)
ax1.plot(available_years, ratios, color=color, label="Labor Force Ratio")
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # zweite y-Achse
color = 'tab:red'
ax2.set_ylabel("TFR", color=color)
ax2.scatter(x, y, s=10, alpha=0.6, label="TFR", color=color)
#ax2.scatter(available_years, sweden_tfr_data, color=color, label="TFR", s=10)
ax2.tick_params(axis='y', labelcolor=color)

plt.title(f"Selected Countries – Female Labor Participation vs TFR ({available_years[0]}–{available_years[-1]})")
plt.xticks(years_every_10, rotation=45)
fig.tight_layout()

plt.savefig(f"diagram_output/selected_countries_tfr_vs_laborforce_dualaxis.png", dpi=1200)
plt.show()
```
