## Import Packages

```python
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
import os

# create output folder
os.makedirs("diagram_output", exist_ok=True)
```

## Load Excel and create diagram

```python
output = pd.read_excel("fertility_rates.xls", sheet_name="Data", index_col=0, skiprows=3)
countries = ["Sweden", "Niger", "China", "Brazil", "Korea, Rep."]
for country in countries:
    print(country)
    country_data = output.loc[country] ###

    # Get x/y values
    years = country_data.index[3:]  # skip first 3 columns ###
    fertility_rates = country_data[3:].values

    # Create diagram
    plt.plot(years, fertility_rates) #, marker='o', markersize=4

    # Configure diagram
    plt.xticks(years[::10], rotation=45)  # Zeige nur jedes 10. Jahr und drehe die Beschriftungen für bessere Lesbarkeit
    plt.title(f"Fertility Rate in {country} (1960–2023)") ###
    plt.xlabel("Year")
    plt.ylabel("Fertility Rate (births per woman)")
    plt.grid(True)
    plt.tight_layout()

    # Save diagram as .png
    plt.savefig(f"diagram_output/{country}_fertility_rate.png", dpi=1200) ###

    # Show diagram
    #plt.show()
```

## Display Sweden Fertility Rate in Comparison to world

```python
countries = ["Sweden", "Niger", "China", "Brazil", "Korea, Rep."]
for country in countries:
    print(country)
    # Read Data
    output = pd.read_excel("fertility_rates.xls", sheet_name="Data", index_col=0, skiprows=3)
    country_data = output.loc[country]
    world_data = output.loc["World"]

    # X-Values (years)
    years = country_data.index[3:]  # erste 3 Spalten überspringen

    # Y-Values
    sweden_rates = country_data[3:].values
    world_rates = world_data[3:].values

    # Create diagram
    plt.plot(years, sweden_rates, label=country)
    plt.plot(years, world_rates, label="World")

    # Configure diagram
    plt.xticks(years[::10], rotation=45)
    plt.title(f"Fertility Rate: {country} vs. World (1960–2023)")
    plt.xlabel("Year")
    plt.ylabel("Fertility Rate (births per woman)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()

    # Save diagram
    plt.savefig(f"diagram_output/{country}_vs_world_fertility_rate.png", dpi=1200)

    # Show diagram
    #plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV einlesen – oder `pd.read_excel(...)` falls es eine Excel-Datei ist
df = pd.read_csv("gini-coefficient-world-bank-marimekko.csv")  # Datei-Pfad hier anpassen

# Spalten vereinheitlichen
df.columns = df.columns.str.strip()

# Länder-Auswahl (deine fünf Länder)
selected_countries = ["Sweden", "Brazil", "China", "Korea, Rep.", "Niger"]

# Filter anwenden
df_filtered = df[df["Entity"].isin(selected_countries)]

# World-Durchschnitt berechnen:
# Wir nehmen alle Länder pro Jahr und bilden den Mittelwert des Gini-Koeffizienten
df_world = df.groupby("Year")["Gini coefficient (2017 prices) - Income or consumption consolidated"].mean().reset_index()

# Plot vorbereiten
plt.figure(figsize=(10, 6))

# Für jedes Land einzeln zeichnen
for country in selected_countries:
    country_data = df_filtered[df_filtered["Entity"] == country]
    plt.plot(
        country_data["Year"],
        country_data["Gini coefficient (2017 prices) - Income or consumption consolidated"],
        label=country,
        #marker='o'
    )

# Welt plot hinzufügen
plt.plot(
    df_world["Year"],
    df_world["Gini coefficient (2017 prices) - Income or consumption consolidated"],
    label="World Average",
    color="black",
    #linestyle="--",
    #linewidth=2
)

# Diagramm-Details
plt.title("Gini-Koeffizient (2017 Preise) – ausgewählte Länder")
plt.xlabel("Jahr")
plt.ylabel("Gini-Koeffizient")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Optional speichern
plt.savefig("diagram_output/gini_selected_countries.png", dpi=300)
#plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV einlesen
df = pd.read_csv("death-rate-from-suicides.csv")  # falls mit Tab getrennt, sonst sep=','

# Spalten umbenennen zur einfacheren Verarbeitung
df.columns = ["Entity", "Code", "Year", "SuicideRate"]

# Länder/Regionen, die verglichen werden sollen
countries = ["Sweden", "Niger", "China", "Brazil", "South Korea"]
comparison_region = "World"  # oder z. B. "Africa", "Asia"

# In float konvertieren (Komma statt Punkt bei einigen Werten?)
df["SuicideRate"] = df["SuicideRate"].astype(str).str.replace(",", ".").astype(float)

# Jahre als Integer
df["Year"] = df["Year"].astype(int)

# Für jedes Land ein Diagramm zeichnen
for country in countries:
    print(f"Erstelle Diagramm für {country}...")

    # Daten extrahieren
    country_data = df[df["Entity"] == country].sort_values("Year")
    world_data = df[df["Entity"] == comparison_region].sort_values("Year")

    # Falls Daten fehlen, überspringen
    if country_data.empty or world_data.empty:
        print(f"Daten fehlen für {country} oder {comparison_region}")
        continue

    # Diagramm erstellen
    plt.figure(figsize=(10, 5))
    plt.plot(country_data["Year"], country_data["SuicideRate"], label=country)
    plt.plot(world_data["Year"], world_data["SuicideRate"], label=comparison_region, linestyle="--")

    plt.title(f"Suizidrate: {country} vs. {comparison_region} (2000–2021)")
    plt.xlabel("Jahr")
    plt.ylabel("Sterberate durch Suizid (altersstandardisiert)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # Speichern
    filename = f"diagram_output/{country}_vs_{comparison_region}_suicide_rate.png"
    plt.savefig(filename, dpi=300)
    print(f"Diagramm gespeichert: {filename}")

    # plt.show() optional
    # plt.show()
    plt.close()
```
