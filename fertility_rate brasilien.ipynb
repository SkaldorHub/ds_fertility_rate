```python
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
import os

# create output folder
os.makedirs("diagram_output", exist_ok=True)
```

```python
output = pd.read_excel("fertility_rates.xls", sheet_name="Data", index_col=0, skiprows=3)
countries = ["Sweden", "Niger", "China", "Brazil", "Korea, Rep."]
for country in countries:
    print(country)
    country_data = output.loc[country] ###

    # Get x/y values
    years = country_data.index[3:]  # skip first 3 columns ###
    fertility_rates = country_data[3:].values

    # Create diagram
    plt.plot(years, fertility_rates) #, marker='o', markersize=4

    # Configure diagram
    plt.xticks(years[::10], rotation=45)  # Zeige nur jedes 10. Jahr und drehe die Beschriftungen fr bessere Lesbarkeit
    plt.title(f"Fertility Rate in {country} (19602023)") ###
    plt.xlabel("Year")
    plt.ylabel("Fertility Rate (births per woman)")
    plt.grid(True)
    plt.tight_layout()

    # Save diagram as .png
    plt.savefig(f"diagram_output/{country}_fertility_rate.png", dpi=1200) ###

    # Show diagram
    #plt.show()
```

## World comparison

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Data
output = pd.read_excel("fertility_rates.xls", sheet_name="Data", index_col=0, skiprows=3)
brazil_data = output.loc["Brazil"]
world_data = output.loc["World"]



# Y-Values

brazil_rates = brazil_data[3:].values
world_rates = world_data[3:].values

# Create diagram

plt.plot(years, brazil_rates, label="Brasilien")
plt.plot(years, world_rates, label="Welt")

# Configure diagram
plt.xticks(years[::10], rotation=45)
plt.title("TFR: Brasilien vs. Welt (19602023)")
plt.xlabel("Jahr")
plt.ylabel("TFR (Geburten pro Frau)")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Save diagram
plt.savefig("diagram_output/world_fertility_rate.png", dpi=1200)

# Show diagram
# plt.show()
```
