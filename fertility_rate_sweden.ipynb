Import Packages

```python
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
```

Load Excel and create diagram

```python
# set max rows to unlimited
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

output = pd.read_excel("fertility_rates.xls", sheet_name="Data", index_col=0, skiprows=3)

# Countries that will be displayed
# countries = ["Sweden", "South Korea", "Niger", "Brazil", "China"]
# country_data = {}

sweden_data = output.loc["Sweden"]

# Get x/y values
years = sweden_data.index[3:]  # skip first 3 rows
fertility_rates = sweden_data[3:].values

# Create diagram
plt.plot(years, fertility_rates, marker='o', markersize=4)

# Configure diagram
plt.xticks(years[::10], rotation=45)  # Zeige nur jedes 10. Jahr und drehe die Beschriftungen für bessere Lesbarkeit
plt.title("Fertility Rate in Sweden (1960–2023)")
plt.xlabel("Year")
plt.ylabel("Fertility Rate (births per woman)")
plt.grid(True)

# Show diagram
#plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
