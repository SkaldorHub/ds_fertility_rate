Import Packages

```python
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
import os

# create output folder
os.makedirs("diagram_output", exist_ok=True)
```

Load Excel and create diagram

```python
output = pd.read_excel("fertility_rates.xls", sheet_name="Data", index_col=0, skiprows=3)
sweden_data = output.loc["Sweden"]

# Get x/y values
years = sweden_data.index[3:]  # skip first 3 columns
fertility_rates = sweden_data[3:].values

# Create diagram
plt.plot(years, fertility_rates) #, marker='o', markersize=4

# Configure diagram
plt.xticks(years[::10], rotation=45)  # Zeige nur jedes 10. Jahr und drehe die Beschriftungen für bessere Lesbarkeit
plt.title("Fertility Rate in Sweden (1960–2023)")
plt.xlabel("Year")
plt.ylabel("Fertility Rate (births per woman)")
plt.grid(True)
plt.tight_layout()

# Save diagram as .png
plt.savefig("diagram_output/sweden_fertility_rate.png", dpi=1200)

# Show diagram
plt.show()
```

Display Sweden Fertility Rate in Comparison to world

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Data
output = pd.read_excel("fertility_rates.xls", sheet_name="Data", index_col=0, skiprows=3)
sweden_data = output.loc["Sweden"]
world_data = output.loc["World"]

# X-Values (years9
years = sweden_data.index[3:]  # erste 3 Spalten überspringen

# Y-Values
sweden_rates = sweden_data[3:].values
world_rates = world_data[3:].values

# Create diagram
plt.plot(years, sweden_rates, label="Sweden")
plt.plot(years, world_rates, label="World")

# Configure diagram
plt.xticks(years[::10], rotation=45)
plt.title("Fertility Rate: Sweden vs. World (1960–2023)")
plt.xlabel("Year")
plt.ylabel("Fertility Rate (births per woman)")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Save diagram
plt.savefig("diagram_output/sweden_vs_world_fertility_rate.png", dpi=1200)

# Show diagram
```

World comparison

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Data
output = pd.read_excel("fertility_rates.xls", sheet_name="Data", index_col=0, skiprows=3)
sweden_data = output.loc["Sweden"]
brazil_data = output.loc["Brazil"]
niger_data = output.loc["Niger"]
china_data = output.loc["China"]
south_korea_data = output.loc["Korea, Rep."]
world_data = output.loc["World"]

# X-Values (years9
years = sweden_data.index[3:]  # erste 3 Spalten überspringen

# Y-Values
sweden_rates = sweden_data[3:].values
brazil_rates = brazil_data[3:].values
niger_rates = niger_data[3:].values
china_rates = china_data[3:].values
south_korea_rates = south_korea_data[3:].values
world_rates = world_data[3:].values

# Create diagram
plt.plot(years, sweden_rates, label="Sweden")
plt.plot(years, brazil_rates, label="Brazil")
plt.plot(years, niger_rates, label="Niger")
plt.plot(years, china_rates, label="China")
plt.plot(years, south_korea_rates, label="South Korea")
plt.plot(years, world_rates, label="World")

# Configure diagram
plt.xticks(years[::10], rotation=45)
plt.title("Fertility Rate: Sweden vs. World (1960–2023)")
plt.xlabel("Year")
plt.ylabel("Fertility Rate (births per woman)")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Save diagram
plt.savefig("diagram_output/world_fertility_rate.png", dpi=1200)

# Show diagram
```
